```python
# Fonction pour cropper les images et récupérer que le visage
def crop_image(folder,image_path):
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    try:
        image = cv2.imread(image_path)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.1, 4)

        if len(faces) > 0:
            for i, (x, y, w, h) in enumerate(faces):
                face_crop = image[y:y+h, x:x+w]
                filename = os.path.basename(image_path)
                cv2.imwrite(os.path.join(folder+'_bis/', filename), face_crop)
        else:
            print(f"Aucun visage détecté dans l'image: {image_path}")
    except Exception as e:
        print(f"Erreur lors du traitement de l'image {image_path}: {e}")
```

```python
# Lancer la fonction pour cropper les images sur les différents dossiers d'images
for folder in ['part1','part2','part3']: 
    os.makedirs(folder+'_bis')
    files = os.listdir(folder)
    for filename in files:
            image_path = os.path.join(folder, filename)
            crop_image(folder,image_path)
```

```python
# Fonction pour compter le nombre d'images dans un dossier
def count_items_in_folder(folder_path):
    if not os.path.exists(folder_path):
        print(f"Le dossier spécifié n'existe pas: {folder_path}")
        return None
    items = os.listdir(folder_path)
    num_items = len(items)

    return num_items
```

```python
for folder in ['part1','part1_bis','part2','part2_bis','part3','part3_bis']: 
    print(folder+' nb images: '+str(count_items_in_folder(folder)))
```

```python
# fonction pour transformer resize l'image et récupérer les labels
def load_and_resize_images(base_dir, size=(224, 224)):
    image_data = []
    labels = []

    # Vérifier si le dossier existe bien
    if not os.path.exists(base_dir):
        raise ValueError(f"Le dossier spécifié n'existe pas: {base_dir}")
    
    # Parcourir chaque fichier dans le dossier
    for filename in os.listdir(base_dir):
        if filename.endswith('.jpg'):
            # Construire le chemin complet vers l'image
            file_path = os.path.join(base_dir, filename)
            
            label_parts = filename.split('_')
            # S'assurer que le fichier a un format de nom correct avant de continuer
            if len(label_parts) != 4 or not all(part.isdigit() for part in label_parts[:3]):
                print(f"Ignorer le fichier non conforme: {filename}")
                continue  # Ignorer ce fichier et passer au suivant
            
            # Charger l'image et la redimensionner
            try:
                image = Image.open(file_path).convert('RGB')
                image = image.resize(size)
            except Exception as e:
                print(f"Impossible de traiter l'image {filename}: {e}")
                continue  # Ignorer ce fichier et passer au suivant

            # Convertir l'image en un tableau numpy et l'ajouter à la liste
            image_np = np.array(image)
            image_data.append(image_np)

            labels.append({
                'age': int(label_parts[0]),
                'gender': int(label_parts[1]),
                'race': int(label_parts[2]),
                'date&time': label_parts[3].split('.')[0] 
            })
            
            image.close()

    return np.array(image_data), labels
```

```python
images1, image_labels1 = load_and_resize_images('part1_bis')
images2, image_labels2 = load_and_resize_images('part2_bis')
images3, image_labels3 = load_and_resize_images('part3_bis')
```

```python
# Concaténer les tableaux d'images
images_concat = np.concatenate((images1, images2, images3), axis=0)

# Concaténer les listes de labels
labels_concat = image_labels1 + image_labels2 + image_labels3
```

```python
# Convertir les étiquettes en DataFrame
labels_df = pd.DataFrame(labels_concat)
labels_df.to_csv('labels_df.csv')
```

```python
# Regrouper les images croppés et resizés dans un seul repo output_images
def save_images(images, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    for i, image in enumerate(images):
        output_path = os.path.join(output_folder, f"image_{i}.jpg")
        cv2.imwrite(output_path, image)
```

```python
# Utilisation de la fonction
output_folder = "output_images"
save_images(images_concat, output_folder)
```
